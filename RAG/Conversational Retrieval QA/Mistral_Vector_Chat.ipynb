# Translation

```python
import translators as ts

SERVICE = "google"


def translate_to_pt(text: str) -> str:
    return ts.translate_text(
        text, from_language="en", to_language="pt", translator=SERVICE
    )


def translate_to_en(text: str) -> str:
    return ts.translate_text(
        text, from_language="pt", to_language="en", translator=SERVICE
    )
```

# LLM Def

```python
from langchain.llms import CTransformers
#'temperature': 0.8

#Seed1 = 1521 -> Resultado ok
config = {'max_new_tokens': 412, 'temperature': 0.8, 'repetition_penalty': 1.1, 'context_length': 2048, 'gpu_layers': 10_000_000_000, 'seed': 1500}

llm = CTransformers(model='TheBloke/Mistral-7B-Instruct-v0.1-GGUF', model_file='mistral-7b-instruct-v0.1.Q6_K.gguf', model_type='mistral', config=config)
```

# Embeddings Def

```python
from langchain.embeddings.sentence_transformer import SentenceTransformerEmbeddings
embedding_function = SentenceTransformerEmbeddings(model_name="BAAI/bge-large-en-v1.5")
```

# Vectorstore Def

```python
from langchain.vectorstores import Chroma

db = Chroma('llm_store', embedding_function)
```

# Adding to the store

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.document_loaders import TextLoader


def add_txt_to_db(filename: str, chunk_size: int, chunk_overlap: int) -> list[str]:
    """Retorna os ids dos documentos adicionados"""

    loader = TextLoader(f"{filename}.txt")
    documents = loader.load()
    text_splitter = RecursiveCharacterTextSplitter(chunk_size=chunk_size, chunk_overlap=chunk_overlap)
    return db.add_documents(text_splitter.split_documents(documents))

def add_pt_txt_to_db(filename: str, chunk_size: int, chunk_overlap: int) -> list[str]:
    """Retorna os ids dos documentos adicionados"""

    loader = TextLoader(f"{filename}.txt")
    documents = loader.load()
    text_splitter = RecursiveCharacterTextSplitter(chunk_size=chunk_size, chunk_overlap=chunk_overlap)
    documents = []
    for doc in text_splitter.split_documents(documents):
      doc.page_content = translate_to_en(doc.page_content)
      documents.append(doc)

    return db.add_documents(documents)

def add_texts_to_db(texts: list[str]) -> list[str]:
    """Retorna os ids dos documentos adicionados"""

    return db.add_texts(texts)

def add_pt_texts_to_db(text: list[str]) -> list[str]:
    """Retorna os ids dos documentos adicionados"""

    translated = []
    for el in text:
      translated.append(translate_to_en(el))
    return add_texts_to_db(translated)

def clear_db():
    db.delete_collection()
```

# Create RetrievalQA Chain

```python
from langchain import PromptTemplate
from langchain.chains import ConversationalRetrievalChain

test_template = """<s>[INST] You are a helpful assistant.
Always call Federal University of Santa Catarina as UFSC.
Use the following context to Answer the question below briefly:
Context: [ {context} ]
Question: {question} [/INST] </s>
"""

standalone_test_question_template = """<s>[INST] You are a helpful assistant.
Summarize the following conversation and a follow up question into a standalone question briefly:
Chat History: [ {chat_history} ]
Follow Up Question: {question} [/INST] </s>
"""


def create_retrieval_chain(k: int = 4,score_threshold: float = 0.0) -> ConversationalRetrievalChain:

    QA_CHAIN_PROMPT = PromptTemplate.from_template(test_template)

    CONDENSE_PROMPT = PromptTemplate.from_template(standalone_test_question_template)

    return ConversationalRetrievalChain.from_llm(
    llm,
    retriever=db.as_retriever(search_kwargs={'k': k, 'score_threshold': score_threshold}),
    condense_question_prompt=CONDENSE_PROMPT,
    combine_docs_chain_kwargs={"prompt": QA_CHAIN_PROMPT},
    verbose = True
)

qa_chain = create_retrieval_chain(3)
```

# Testing and Generating Text

## Generation

```python
def generate(query: str, chat_history) -> str:
    return qa_chain({'question': query, 'chat_history': chat_history})

def testing_input(st: list[str]):
    chat_history = []

    for query in st:
      print('Q:', query)
      output = generate(query, chat_history)
      print('A:', output['answer'])
      chat_history.append((query, output['answer']))

def testing_input_pt(st: list[str]):
    chat_history = []

    for query in st:
      print('Pergunta:', query)
      translated_query = translate_to_en(query)
      output = generate(translated_query, chat_history)
      translated_output = translate_to_pt(output['answer'])
      print('Resposta:', translated_output)
      chat_history.append((translated_query, output['answer']))
```
